```python
%matplotlib inline
import mdtraj as md
import mixtape.featurizer, mixtape.cluster, mixtape.tica
import sklearn.pipeline
import glob
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ACTIVE = md.load_pdb('../../../exacycleGPCR/GPCR_NatureChemistry/reference-structures/active_crystal_reference.pdb')
INACTIVE = md.load('../../../exacycleGPCR/GPCR_NatureChemistry/reference-structures/inactive_crystal_reference.pdb')
APO = md.load('../../../exacycleGPCR/GPCR_NatureChemistry/reference-structures/apo_snapshot.pdb')
ACTIVE_union = md.load('../../../exacycleGPCR/GPCR_NatureChemistry/reference-structures/active-union.pdb')
```

```python
filenames = glob.glob('../../../exacycleGPCR/dcd_trajectories/apo_b2ar/trj11*')
traj = [md.load(filename, top=APO) for filename in filenames]
table, bonds = traj[1].topology.to_dataframe()

# atom selections
I121_F282 = table[(table['resSeq'] == 121) | (table['resSeq'] == 282)].index  
NPxxY = {'resSeq': [322,323,324,325,326,327]}
NPxxY_region = table[table.isin(NPxxY).any(1)].index
R131_L272 = table[(table['resSeq'] == 131) & (table['name'] == 'CA') | (table['resSeq'] == 272) & (table['name'] == 'CA')].index
binding_pocket = {'resSeq': [109, 110, 113, 114, 117, 118, 191, 192, 193, 195, 199, 200, 203, 204, 207, 286, 289, 290, 293, 305, 308, 309, 312, 316]}
binding_pocket_RMSD = table[table.isin(binding_pocket).any(1)].index
```

```python
distance = np.array([R131_L272])
Hel3_Hel6_dist = mixtape.featurizer.AtomPairsFeaturizer(distance)
NPxxY_RMSD = mixtape.featurizer.RMSDFeaturizer(ACTIVE_union, NPxxY_region)
Connector_RMSD = mixtape.featurizer.RMSDFeaturizer(ACTIVE_union, I121_F282)
bind_pock_RMSD = mixtape.featurizer.RMSDFeaturizer(ACTIVE_union, binding_pocket_RMSD)
```

```python
union = mixtape.featurizer.TrajFeatureUnion([("Helical_distance", Hel3_Hel6_dist), ("NPxxYrmsd", NPxxY_RMSD), ("connecector", Connector_RMSD), ("binding pocket", bind_pock_RMSD)])
combined_features = union.fit_transform(traj)
```

```python
Hel_distance = [union.transformer_list[0][-1].partial_transform(t) for t in traj]
Hel_distance_all = np.concatenate(Hel_distance)
plt.hist(Hel_distance_all, bins=50)
```

```python
NPxxY_rmsd = [union.transformer_list[1][-1].partial_transform(t) for t in traj]
NPxxY_rmsd_all = np.concatenate(NPxxY_rmsd)
plt.hist(NPxxY_rmsd_all, bins=50)
```

```python
connector_rmsd = [union.transformer_list[2][-1].partial_transform(t) for t in traj]
connector_rmsd_all = np.concatenate(connector_rmsd)
plt.hist(connector_rmsd_all, bins=50)
```

```python
n_components = 3
n_states = 3
lag_time = 1
tica = mixtape.tica.tICA(n_components=n_components, lag_time=lag_time)
cluster = mixtape.cluster.GMM(n_components=n_states, covariance_type='diag')
feature_pipeline = sklearn.pipeline.Pipeline([('features', union), ('tica', tica)])
cluster_pipeline = sklearn.pipeline.Pipeline([('features', union), ('tica', tica), ('cluster', cluster)])
without_tica = sklearn.pipeline.Pipeline([('features', union), ('cluster', cluster)])
```

```python
X_all = feature_pipeline.fit_transform(traj)
q = np.concatenate(X_all)
```
